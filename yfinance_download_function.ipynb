```python
import yfinance as yf
import pandas as pd
# Set global float format to two decimal places
pd.options.display.float_format = '{:.2f}'.format
```

# Download single ticker

```python
start_date = pd.to_datetime("2020-01-01")
end_date = pd.to_datetime("today")

tesla_data = yf.download("TSLA", start=start_date, end=end_date)
tesla_data.index = tesla_data.index.strftime('%Y-%m-%d')
```

```python
tesla_data.head()
```

## Remove multi level index


### Method 1: Join the multi-level with an underscore (_)

```python
df_copy = tesla_data.copy()
df_copy.columns = ['_'.join(col).strip() for col in df_copy.columns.values]
display(df_copy.head())
```

### Method 2: Reset columns to default numeric index

```python
df_copy.columns = df_copy.columns.to_flat_index()
display(df_copy.head())
```

### Method 3: Drop the unwanted column index

```python
# Drop the first level (level=0)
df_copy = tesla_data.copy()
df_copy.columns = df_copy.columns.droplevel(1)
display(df_copy.head())
```

```python
# Or drop by name if the levels are named
df_copy = tesla_data.copy()
df_copy.columns = df_copy.columns.droplevel('Ticker')
display(df_copy.head())
```

## Multi-Tickers with a Single DataFrame

```python
import yfinance as yf
import pandas as pd
import time

# Define tickers and time period
tickers = ["TSLA", "AAPL", "MSFT"]
start_date = pd.to_datetime("2020-01-01")
end_date = pd.to_datetime("today")

# Start timing
start_time = time.time()

# Download all data
data = yf.download(tickers, start=start_date, end=end_date)
data.index = data.index.strftime('%Y-%m-%d')

# Split into individual DataFrames for each ticker
df_list = {ticker: data.xs(ticker, level=1, axis=1) for ticker in tickers}

# Stop timing
end_time = time.time()
print(f"\nMethod 1 runtime: {end_time - start_time:.2f} seconds")

# Example: Display Tesla's DataFrame
display(df_list["TSLA"].head())
```

```python
display(df_list)
```

```python
# Start timing
start_time = time.time()

# Initialize an empty dictionary to store individual DataFrames
df_list = {}

# Download data for each ticker separately
for ticker in tickers:
    df = yf.download(ticker, start=start_date, end=end_date)
    df.index = df.index.strftime('%Y-%m-%d')
    df_list[ticker] = df

# Stop timing
end_time = time.time()
print(f"\nMethod 2 runtime: {end_time - start_time:.2f} seconds")

# Example: Display Tesla's DataFrame
display(df_list["TSLA"].head())
```
